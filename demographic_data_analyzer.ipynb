```python
demographic_data = demographic_data.rename(columns={"education-num":"total_years",
                                                   "marital-status":"status",
                                                    "capital-gain" :"gain",
                                                    "capital-loss" :"loss",
                                                    "hours-per-week":"working_hours",
                                                    "native-country":"homeland"
                                                   })

# print dataframe columns
print("Dataframe columns:", demographic_data.columns)
```

```python
# Remove special character '-' from the data set except "education" column
demographic_data['age']= demographic_data['age'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['workclass']= demographic_data['workclass'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['fnlwgt']= demographic_data['fnlwgt'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['total_years']= demographic_data['total_years'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['status']= demographic_data['status'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['occupation']= demographic_data['occupation'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['relationship']= demographic_data['relationship'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['race']= demographic_data['race'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['sex']= demographic_data['sex'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['gain']= demographic_data['gain'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['loss']= demographic_data['loss'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['working_hours']= demographic_data['working_hours'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['homeland']= demographic_data['homeland'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
demographic_data['salary']= demographic_data['salary'].apply(lambda x: str(x).replace('-',' ')if '-' in str(x) else str(x))
```

```python
# merge all the rows which have '?' by using or
print(demographic_data[demographic_data.workclass.str.contains('[?]')
         | demographic_data.education.str.contains(r'[?]') 
         | demographic_data.status.str.contains(r'[?]') 
         | demographic_data.occupation.str.contains(r'[?]') 
         | demographic_data.relationship.str.contains(r'[?]')
         | demographic_data.race.str.contains(r'[?]')
         | demographic_data.sex.str.contains(r'[?]') 
         | demographic_data.homeland.str.contains(r'[?]') ])
```

```python
# drop the merged selected rows
df=demographic_data.drop(demographic_data[demographic_data.workclass.str.contains('[?]')
         | demographic_data.education.str.contains(r'[?]') 
         | demographic_data.status.str.contains(r'[?]') 
         | demographic_data.occupation.str.contains(r'[?]') 
         | demographic_data.relationship.str.contains(r'[?]')
         | demographic_data.race.str.contains(r'[?]')
         | demographic_data.sex.str.contains(r'[?]') 
         | demographic_data.homeland.str.contains(r'[?]') ].index)
df.reset_index(drop=True, inplace=True)
```
